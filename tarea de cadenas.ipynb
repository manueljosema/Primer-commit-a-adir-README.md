```python
try:
    edad = int(input("¿Cuántos años tienes? "))
    print(f"Tienes {edad} años.")
except ValueError:
    print("Por favor, ingresa un número válido.")
```

```python
nombre_completo = "Juan Pérez"
nombre_mayusculas = nombre_completo.upper()
nombre_minusculas = nombre_completo.lower()
print(nombre_mayusculas)
print(nombre_minusculas)



```

```python
total_caracteres = len(nombre_completo)
print(total_caracteres)
```

```python
solo_letras = len(nombre_completo.replace(" ", ""))
print(solo_letras)
```

```python
nombre_modificado = nombre_completo.replace("Juan", "Carlos")
print(nombre_modificado)
```

```python
nombres = input("Ingresa una lista de nombres separados por comas: ")
lista_nombres = nombres.split(",")

for nombre in lista_nombres:
    print(nombre.strip())  # Usamos .strip() para eliminar espacios extra
```
