```python
# Verificar si hay duplicados y borrarlos.
df.drop_duplicates(inplace=True)

# todo ok
```

```python
df = pd.DataFrame(df)

# Verificar los tipos de datos de cada columna
print(df.dtypes)

#totalcharges estaba en formato string y lo cambie a float
```

```python
# Reemplazar espacios con NaN
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)

# Cambiar el tipo de dato a float64
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Verificar el tipo de dato después de cambiar
print(df.dtypes)
```

No hay errores ortograficos en referencia a:
> Identificar y solucionar cualquier inconsistencia en los valores categóricos (ejemplo: cat, Cat, cats).



```python
#codigo para identificar potenciales inconsistencias
for column in df.select_dtypes(include=['object']).columns:
    print(f"Unique values for {column}:")
    print(df[column].unique())
    print("\n")
```

```python
# Utiliza el método isnull().sum() para contar valores faltantes por columna.
print(df.isnull().sum())

#Solucionar valores faltantes
#Rellenar con mediana:
df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
#Utilizaré el método value_counts() para detectar valores únicos.
print(df.nunique())
```

###**2- Exploración de datos con Python:**
* Crear dos tipos diferentes de visualizaciones exploratorias univariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.
* Crear dos tipos diferentes de visualizaciones exploratorias multivariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.


```python
# Calcular promedio
promedio = df['MonthlyCharges'].mean()

# Histograma de 'MonthlyCharges'
plt.figure(figsize=(8, 6))
sns.histplot(df['MonthlyCharges'], bins=20, kde=True)

# Agregar línea vertical para el promedio
plt.axvline(promedio, color='red', linestyle='--', label=f'Promedio: ${promedio:.2f}')

# Mostrar el promedio en el título
plt.title(f'Distribución de Cargos Mensuales (Promedio: ${promedio:.2f})')

plt.xlabel('Cargos Mensuales ($)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

# Interpretación:
# La distribución de cargos mensuales es aproximadamente normal,
# con la mayoría de los valores entre $30 y $100.
```

```python
# Mostrar estadísticas resumen
print("\nEstadísticas resumen:")
print(df['MonthlyCharges'].describe())
```

```python
# Gráfico de barras para 'gender'
plt.figure(figsize=(8,6))
sns.countplot(x='gender', data=df)

# Agregar porcentajes
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width()/2, p.get_height(),
             f"{p.get_height()/len(df)*100:.1f}%", ha="center", va="bottom")

plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.show()

# Interpretación:
# La mayoría de los clientes son hombres (aproximadamente 51%),
# Mientras que las mujeres representan alrededor del 50%.
# La diferencia oscila en 1% por ende no hay gran impacto en la distribución de genero.
```

```python
# Scatterplot de 'Cargos Mensuales' vs 'Tenencia'
plt.figure(figsize=(8,6))
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Relación entre Tenencia y Cargos Mensuales')
plt.xlabel('Tenencia (meses)')
plt.ylabel('Cargos Mensuales ($)')
plt.legend(title='Churn')
plt.show()

# Interpretación:
# La relación entre tenencia y cargos mensuales es positiva,
# indicando que clientes con mayor tenencia tienden a tener mayores cargos.
```

```python
# Gráfico de barras apiladas de 'Contratos' y 'Metodos de Pago'
plt.figure(figsize=(8,6))
df.pivot_table(index='Contract', columns='PaymentMethod', aggfunc='size', fill_value=0).plot(kind='bar', stacked=True)
plt.title('Distribución de Contratos y Métodos de Pago')
plt.xlabel('Contrato')
plt.ylabel('Frecuencia')
plt.legend(title='Método de Pago')
plt.show()

# Interpretación:
# La mayoría de los clientes tienen contratos de 1 año
# y utilizan **tarjetas de crédito** y **transferencias de banco**
# para el pago por una cuestión de practicidad.
```
