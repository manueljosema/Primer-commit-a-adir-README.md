###2) **Crear Histogramas:**
* Crea histogramas para las variables Score, GDP per capita, Healthy life expectancy y Freedom to make life choices.
* Personaliza los histogramas con colores, bordes y ajustes de transparencia.

```python
# Crear histogramas
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i, var in enumerate(variables):
    sns.histplot(df[var], ax=axs[i//2, i%2], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    axs[i//2, i%2].set_title(var)
    axs[i//2, i%2].set_xlabel('Valor')
    axs[i//2, i%2].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

### 3) **Crear Boxplots:**
* Crea boxplots para las mismas variables para visualizar la mediana, el rango intercuartil y los valores atípicos.
* Personaliza los boxplots con diferentes colores y estilos.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i, var in enumerate(variables):
    sns.boxplot(df[var], ax=axs[i//2, i%2], color='lightgreen')
    axs[i//2, i%2].set_title(var)
    axs[i//2, i%2].set_xlabel('')
    axs[i//2, i%2].set_ylabel('Valor')

plt.tight_layout()
plt.show()
```

### 3) **Análisis Combinado:**
* Combina los histogramas y los boxplots en una sola figura para cada variable.
* Interpreta los resultados y escribe un análisis sobre la distribución de cada variable y la presencia de valores atípicos.

```python
# Análisis combinado
fig, axs = plt.subplots(4, 2, figsize=(12, 16))

for i, var in enumerate(variables):
    sns.histplot(df[var], ax=axs[i, 0], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    axs[i, 0].set_title(var + ' - Histograma')
    axs[i, 0].set_xlabel('Valor')
    axs[i, 0].set_ylabel('Frecuencia')

    sns.boxplot(df[var], ax=axs[i, 1], color='lightgreen')
    axs[i, 1].set_title(var + ' - Boxplot')
    axs[i, 1].set_xlabel('')
    axs[i, 1].set_ylabel('Valor')

plt.tight_layout()
plt.show()
```

**Distribución del Puntaje de Felicidad**

El histograma del puntaje de felicidad muestra una distribución aproximadamente normal, con la mayoría de los países concentrados entre 4 y 7 puntos. Esto sugiere que la felicidad es relativamente consistente entre los países. Sin embargo, el boxplot revela algunos valores atípicos con puntajes muy altos (superiores a 8) o muy bajos (inferiores a 3), lo que indica países con niveles de felicidad extremadamente altos o bajos.

**Distribución del PIB per Cápita**

El histograma del PIB per cápita muestra una distribución sesgada hacia la derecha, indicando que la mayoría de los países tienen un PIB per cápita bajo. Esto se debe a la presencia de países en desarrollo con economías emergentes. El boxplot confirma la presencia de valores atípicos en la cola derecha de la distribución, lo que sugiere que algunos países tienen PIB per cápita significativamente más alto que la media.

**Distribución de la Expectativa de Vida Saludable**

El histograma de la expectativa de vida saludable muestra una distribución relativamente simétrica con un pico alrededor de los 65 años. Esto indica que la mayoría de los países tienen expectativas de vida saludable similares. El boxplot muestra algunos valores atípicos en ambos extremos, lo que sugiere variabilidad en la expectativa de vida saludable entre diferentes países.

**Distribución de la Libertad para Tomar Decisiones de Vida**

El histograma de la libertad para tomar decisiones de vida muestra una distribución sesgada hacia la derecha, similar a la del PIB per cápita. Esto indica que la mayoría de los países tienen niveles moderados de libertad. El boxplot revela varios valores atípicos, lo que sugiere que en algunos países la libertad para tomar decisiones de vida es significativamente diferente de la mayoría.

**Conclusiones**

- La felicidad y la expectativa de vida saludable muestran distribuciones relativamente consistentes.
- El PIB per cápita y la libertad para tomar decisiones de vida muestran distribuciones sesgadas.
- Valores atípicos en todas las variables indican variabilidad significativa entre países.
- Los resultados sugieren que factores económicos, sociales y políticos influyen en la felicidad y bienestar de los países.

```python
# Interpretación
print("Distribución de variables:")
for var in variables:
    print(f"{var}: {df[var].describe()}")

print("\nValores atípicos:")
for var in variables:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    valores_atipicos = df[var][(df[var] < Q1 - 1.5 * IQR) | (df[var] > Q3 + 1.5 * IQR)]
    print(f"{var}: {valores_atipicos.shape[0]} valores atípicos")
```
