```python
# Assuming you want to create a new column named 'Adjusted Value' which is the 'Value (Actual)' multiplied by 2
# Replace this with your desired operations and column names

# Check if 'Value (Actual)' column exists
if 'Value (Actual)' in df.columns:
    # Create the new 'Adjusted Value' column
    df['Adjusted Value'] = df['Value (Actual)'] * 2

    # Display the first 5 rows of the updated dataframe
    print(df.head())
else:
    print("'Value (Actual)' column not found in the dataframe")
```

```python
# Import necessary libraries
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Select columns for normalization/standardization
numerical_cols = ['Year', 'Value (Actual)', 'Adjusted Value']

# Normalize using MinMaxScaler
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Standardize using StandardScaler (optional)
# scaler = StandardScaler()
# df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
# Convert 'Value (Actual)' to numeric, handling errors
df['Value (Actual)'] = pd.to_numeric(df['Value (Actual)'], errors='coerce')

# Categorize data based on relevant columns

# Categorization based on "?Format"
format_categories = df.groupby("?Format")["Value (Actual)"].sum()
print(format_categories)


# Categorization based on "Metric"
metric_categories = df.groupby("Metric")["Value (Actual)"].sum()
print(metric_categories)

# Categorization based on "Year" (e.g., decades)
df['Decade'] = (df['Year'] // 10) * 10
decade_categories = df.groupby("Decade")["Value (Actual)"].sum()
print(decade_categories)

# Example of creating a new 'Category' column based on multiple criteria
#  (Replace with your specific business logic)

def categorize_data(row):
    if row["?Format"] == "CD" and row["Metric"] == "Value":
        return "CD Sales"
    elif row["?Format"] == "Music Video (Physical)" :
        return "Physical Music Videos"
    else:
        return "Other"

df["Category"] = df.apply(categorize_data, axis=1)
category_sums = df.groupby("Category")["Value (Actual)"].sum()
category_sums
```

```python
# Group data by format and calculate the total value
format_value = df.groupby("?Format")["Value (Actual)"].sum()
print("Total Value by Format:\n", format_value)

# Group data by year and calculate the average value
year_avg_value = df.groupby("Year")["Value (Actual)"].mean()
print("\nAverage Value by Year:\n", year_avg_value)

# Group data by category and decade and count the number of records
category_decade_counts = df.groupby(["Category", "Decade"])["Number of Records"].count()
print("\nNumber of Records by Category and Decade:\n", category_decade_counts)
```

```python
# Group data by '?Format' and apply aggregation functions
aggregated_df = df.groupby('?Format')['Value (Actual)'].agg(['sum', 'mean', 'count', 'min', 'max', 'std', 'var'])

# Display the aggregated data
aggregated_df
```

```python
# Define a function to categorize 'Value (Actual)'
def categorize_value(value):
    if value < 100:
        return 'Low'
    elif value < 500:
        return 'Medium'
    else:
        return 'High'

# Apply the function to create a new column 'Value Category'
df['Value Category'] = df['Value (Actual)'].apply(categorize_value)
# Display the first few rows to check the results
df.head(10)
```
